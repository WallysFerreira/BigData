Extrai a area total de cada bairro no banco de dados

```python
import urllib
import json
import csv

url = 'http://dados.recife.pe.gov.br/api/3/action/datastore_search?resource_id=18e58d3b-8096-4bac-bc18-273bacd7d01c'
fileobj = urllib.request.urlopen(url)
json = json.loads(fileobj.read().decode('utf-8'))
csvoutput = open("../data/bairros_com_mais_area_verde.csv", 'w', newline='')
w = csv.writer(csvoutput)
w.writerow(["bairro", "area total"])

dados = json["result"]["records"]
soma = {}

print(dados)
for record in dados:
    soma[record["nome_bairro"]] = 0

for record in dados:
    soma[record["nome_bairro"]] += float(record["area"].replace(",", "."))

for item in soma:
    w.writerow([item] + [soma[item]])
```
